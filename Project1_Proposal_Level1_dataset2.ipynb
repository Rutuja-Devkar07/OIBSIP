```python
# Basic descriptive statistics for numerical columns
print(data.describe())

# Specific statistics
mean_calories = data['Calories'].mean()
median_calories = data['Calories'].median()
mode_calories = data['Calories'].mode()[0]
std_calories = data['Calories'].std()

print(f"Mean: {mean_calories}, Median: {median_calories}, Mode: {mode_calories}, Standard Deviation: {std_calories}")
```

###data visualization 

```python
# Histogram for Calories
plot.figure(figsize=(6, 6))
sns.histplot(data['Calories'], kde=True, bins=20)
plot.title('Distribution of Calories')
plot.xlabel('Calories')
plot.ylabel('Frequency')
plot.show()
```

```python
plot.figure(figsize=(8, 6))
sns.scatterplot(x='Total Fat', y='Calories', data=data)
plot.title('Total Fat vs Calories')
plot.xlabel('Total Fat (g)')
plot.ylabel('Calories')
plot.show()
```

```python
# Total Calories per Category
category_sales = data.groupby('Category')['Calories'].sum().sort_values(ascending=False)

plot.figure(figsize=(10, 6))
sns.barplot(x=category_sales.index, y=category_sales.values)
plot.title('Total Calories by Category')
plot.xticks(rotation=45)
plot.ylabel('Total Calories')
plot.show()

# Top 10 items with the highest calories
top_items = data[['Item', 'Calories']].sort_values(by='Calories', ascending=False).head(10)
print(top_items)
```
